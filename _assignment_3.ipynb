```python
# Data Cleaning
# Remove all duplicate observations
covid_dataset.duplicated()
```

```python
covid_dataset = covid_dataset.drop_duplicates()
```

```python
# Find missing values in all columns
covid_dataset.isna().sum()
```

```python
#  Remove all observations where continent column value is missing
covid_dataset['continent'].isna()
```

```python
covid_dataset=covid_dataset.dropna(subset=["continent"])
```

```python
covid_dataset['continent'].isna()
```

```python
covid_dataset["continent"].isna().sum()
```

```python
# Fill all missing values with 0
covid_dataset.fillna(0)
```

```python
# Date time format
# Convert date column in datetime format using pandas.to_datetime
covid_dataset["date"]=pd.to_datetime(covid_dataset["date"])
```

```python
covid_dataset['date']
```

```python
# Create new column month after extracting month data from date column
covid_dataset["month"]=pd.DatetimeIndex(covid_dataset["date"]).month
```

```python
covid_dataset["month"]
```

```python
# Find max value in all columns using groupby function on 'continent' column
covid_dataset.groupby("continent").max().reset_index()
```

```python
covid_dataset
```

```python
# Store the result in a new dataframe named 'df_groupby'. (Use df_groupby dataframe for all further analysis
df_groupby = covid_dataset
```

```python
df_groupby
```

```python
# Feature Engineering
#  Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'ArithmeticError
df_groupby["total_deaths_to_total_cases"]=df_groupby["total_deaths"]/df_groupby["total_cases"]
df_groupby
```

```python
# Data Visualization
#  Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
sns.distplot(df_groupby["gdp_per_capita"], kde=False, color="red", bins=30)
```

```python
#  Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
sns.scatterplot(x="gdp_per_capita", y="total_cases",
               data=df_groupby)
```

```python
# Plot Pairplot on df_groupby dataset
sns.pairplot(data = df_groupby,hue = 'continent')
plt.show()
```

```python
# Plot a bar plot of 'continent' column with 'total_cases
sns.barplot(x="continent",
             y="total_cases",
              kind="bar",
             data=df_groupby)
```
